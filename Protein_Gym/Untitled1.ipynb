```python
import matplotlib.pyplot as plt
import pandas as pd

# Read in the data
data = pd.read_csv('/Users/sophiebosseva/Desktop/BRCA1_HUMAN_Findlay_2018.csv')

# Create a dictionary to map mutant names to their respective sequences
mutant_seq = dict(zip(data['mutant'], data['mutated_sequence']))

# Create a list of unique amino acids
aa_list = list(set(''.join(mutant_seq.values())))

# Create a dictionary to map amino acids to their index in the list
aa_dict = {aa_list[i]: i for i in range(len(aa_list))}

# Create the heatmap
heatmap = []
for mutant in data['mutant']:
    seq = mutant_seq[mutant]
    row = [aa_dict[seq[i]] if seq[i] in aa_dict else -1 for i in range(len(seq))]
    heatmap.append(row)

plt.imshow(heatmap, cmap='viridis')
plt.colorbar()
plt.xlabel('Position')
plt.ylabel('Mutant')
plt.title('Sequence-Function Map')
plt.show()

```

```python
import matplotlib.pyplot as plt

# extract the first 5 mutants and their DMS scores
subset_df = df.head(5)[['mutant', 'DMS_score']]

# create a dictionary mapping each mutant name to its DMS score
mutant_scores = dict(zip(subset_df['mutant'], subset_df['DMS_score']))

# define the amino acid sequence for each mutant
sequences = {
    'M1V': 'VDLSALRVEEVQNVINAMQKILECPICLELIKEPVSTKCDHIFCKFLPIPDCKVDKIFRNYQVKRKTGTVLFSLGLLPNIQAVLLPKA',
    'M1T': 'TDLSALRVEEVQNVINAMQKILECPICLELIKEPVSTKCDHIFCKFLPIPDCKVDKIFRNYQVKRKTGTVLFSLGLLPNIQAVLLPKA',
    'M1R': 'RDLSALRVEEVQNVINAMQKILECPICLELIKEPVSTKCDHIFCKFLPIPDCKVDKIFRNYQVKRKTGTVLFSLGLLPNIQAVLLPKA',
    'M1L': 'LDLSALRVEEVQNVINAMQKILECPICLELIKEPVSTKCDHIFCKFLPIPDCKVDKIFRNYQVKRKTGTVLFSLGLLPNIQAVLLPKA',
    'M1K': 'KDLSALRVEEVQNVINAMQKILECPICLELIKEPVSTKCDHIFCKFLPIPDCKVDKIFRNYQVKRKTGTVLFSLGLLPNIQAVLLPKA'
}

# create a list of the amino acids at each position in the sequence
amino_acids = [sequences[m][i] for i in range(len(sequences['M1V']))]

# create the sequence-function map
fig, ax = plt.subplots(figsize=(15, 5))
heatmap = ax.pcolor([list(mutant_scores.values())], cmap='RdBu', vmin=-3, vmax=3)
cbar = plt.colorbar(heatmap)
cbar.ax.set_ylabel('DMS score', rotation=270, labelpad=15)
ax.set_xticks([i + 0.5 for i in range(len(amino_acids))])
ax.set_xticklabels(amino_acids)
ax.set_yticks([0.5])
ax.set_yticklabels(['DMS score'])
for i, mutant in enumerate(mutant_scores):
    ax.text(i + 0.5, 0.5, mutant, ha='center', va='center', fontsize=12)
plt.show()
```

```python
import pandas as pd

# Load data
df = pd.read_csv('data.csv')

# Get list of all amino acids in the mutated_sequence column
amino_acids = list(set(''.join(df['mutated_sequence'])))

# Loop through each amino acid
for amino_acid in amino_acids:
    # Loop through each position
    for position in range(1, len(df['mutated_sequence'][0])):
        # Filter to mutations of interest
        df_mutations = df[df['mutant'].str.startswith(f"{amino_acid}{position}")]
        # Check if all mutations have DMS_score_bin of 0
        if df_mutations['DMS_score_bin'].min() == 0:
            print(f"All mutations of {amino_acid}{position} have a DMS_score_bin of 0.")
```
